```python
from pathlib import Path
import numpy as np 
import pandas as pd
import os
```

```python
direc = Path(r'C:\Users\Krul\Documents\GitHub\cell-painting-dashboard\data\processed')
# direc
filepaths = list(direc.glob(r'**/*.tiff'))
# filepaths
FileNames = list(map(lambda x: os.path.split(x)[1],filepaths))
filepaths = pd.Series(filepaths, name='FilePaths').astype(str)
Labels = pd.Series(FileNames, name='FileNames').astype(str)
img_df = pd.merge(filepaths, Labels, right_index = True, left_index = True)
img_df
```

```python
img_df = pd.merge(filepaths, Labels, right_index = True, left_index = True)
img_df['Row'] = img_df['FileNames'].apply(lambda x: x[1:3]).apply(lambda x: chr(int(x)+64))
img_df['Column'] = img_df['FileNames'].apply(lambda x: x[4:6])
img_df['F'] = img_df['FileNames'].apply(lambda x: x[7:9]) 
img_df['Well'] = img_df['Row'] + img_df['Column']
for aa in ['ch1', 'ch2', 'ch3', 'ch4']:
    img_df['FileNames'] = img_df['FileNames'].apply(lambda x: x.replace(aa,''))

img_df = img_df.drop('FilePaths', axis=1)
img_df
```

```python
img_df = img_df.drop_duplicates().reset_index(drop=True)
img_df
```
